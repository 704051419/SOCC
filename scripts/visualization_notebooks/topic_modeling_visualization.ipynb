```python
import gensim, sys
import pandas as pd
import pyLDAvis.gensim
```

```python
from gensim import corpora, models
from collections import defaultdict
```

```python
df = pd.read_csv('/Users/vkolhatk/Data/SOCC/raw/gnm_articles.csv')
```

```python
df.shape
```

```python
df.columns
```

```python
doc_complete = df['article_text'].tolist()
```

```python
doc_collection = []
for doc in doc_complete:
    doc = doc.replace(r'<p>',' ')
    doc = doc.replace(r'</p>',' ')
    ldoc = doc.split()
    doc_collection.append(ldoc)
```

```python
dictionary = corpora.Dictionary(doc_collection)
dictionary.save('dictionary.dict')
```

```python
dictionary = corpora.Dictionary.load("dictionary.dict")
doc_term_matrix = [dictionary.doc2bow(doc) for doc in doc_collection]
corpus_tfidf = models.TfidfModel(doc_term_matrix)[doc_term_matrix]
```

```python
lda = gensim.models.LdaModel(corpus_tfidf, num_topics=5, passes=50, id2word=dictionary, minimum_probability=0.1)
lda.save('topic.model')
```

```python
print(lda.print_topics(num_topics=5, num_words=10))
```

```python
corpora.MmCorpus.serialize('gnm_data.mm', doc_term_matrix)
c = gensim.corpora.MmCorpus('gnm_data.mm')
```

```python
data = pyLDAvis.gensim.prepare(lda, c, dictionary)
```
